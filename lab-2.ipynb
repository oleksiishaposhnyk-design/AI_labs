```python
from google.colab import files
import pandas as pd

# Завантажуємо файл вручну
uploaded = files.upload()  # з'явиться кнопка "Choose File"

# Читаємо датасет
melbourne_data = pd.read_csv('melb_data.csv')
melbourne_data.describe()
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
print("Розмір після очищення:", melbourne_data.shape)
```

```python
y = melbourne_data.Price
print(y.head())
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

melbourne_model = DecisionTreeRegressor(random_state=1)
melbourne_model.fit(X, y)
print("Модель навчена!")
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

melbourne_model2 = DecisionTreeRegressor()
melbourne_model2.fit(train_X, train_y)

val_predictions = melbourne_model2.predict(val_X)
print("MAE:", mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    return mean_absolute_error(val_y, preds_val)

for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes} \t MAE: {my_mae:.0f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print("Random Forest MAE:", mean_absolute_error(val_y, melb_preds))
```
